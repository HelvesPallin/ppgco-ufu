```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

1. Abrir as imagens coloridas e mostrar a imagem e seus histogramas (separados) cada canal de cor. Considerar os seguintes sistemas de cores:
a. RGB b. HSV ou HSI c. Lab

```python
img1 = cv2.imread("img/image.jpg", cv2.IMREAD_COLOR)
img1RGB = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img1R = img1RGB*[1,0,0]
img1RHist = np.histogram(img1R, bins=255, range=(1,255))[0]
img1G = img1RGB*[0,1,0]
img1GHist = np.histogram(img1G, bins=255, range=(1,255))[0]
img1B = img1RGB*[0,0,1]
img1BHist = np.histogram(img1B, bins=255, range=(1,255))[0]

img2 = cv2.imread("img/image2.jpg", cv2.IMREAD_COLOR)
img2RGB = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

img2R = img2RGB*[1,0,0]
img2RHist = np.histogram(img2R, bins=255, range=(1,255))[0]
img2G = img2RGB*[0,1,0]
img2GHist = np.histogram(img2G, bins=255, range=(1,255))[0]
img2B = img2RGB*[0,0,1]
img2BHist = np.histogram(img2B, bins=255, range=(1,255))[0]

img3 = cv2.imread("img/Lenna.jpg", cv2.IMREAD_COLOR)
img3RGB = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

img3R = img3RGB*[1,0,0]
img3RHist = np.histogram(img3R, bins=255, range=(1,255))[0]
img3G = img3RGB*[0,1,0]
img3GHist = np.histogram(img3G, bins=255, range=(1,255))[0]
img3B = img3RGB*[0,0,1]
img3BHist = np.histogram(img3B, bins=255, range=(1,255))[0]

f, axarr = plt.subplots(6,4, figsize = (18,24))
axarr[0,0].imshow(img1RGB)
axarr[0,0].title.set_text("Imagem Original")
axarr[0,1].imshow(img1R)
axarr[0,1].title.set_text("Imagem Canal R")
axarr[0,2].imshow(img1G)
axarr[0,2].title.set_text("Imagem Canal G")
axarr[0,3].imshow(img1B)
axarr[0,3].title.set_text("Imagem Canal B")
axarr[1,0].bar(np.arange(255), img1RHist, width=1, bottom=None, align='center', label='Histogram', color='red', alpha=0.5)
axarr[1,0].bar(np.arange(255), img1GHist, width=1, bottom=None, align='center', label='Histogram', color='green', alpha=0.4)
axarr[1,0].bar(np.arange(255), img1BHist, width=1, bottom=None, align='center', label='Histogram', color='blue', alpha=0.3)
axarr[1,0].title.set_text("Histograma")
axarr[1,0].set_xlabel("Intensidade")
axarr[1,0].set_ylabel("Ocorrencias")
axarr[1,1].bar(np.arange(255), img1RHist, width=1, bottom=None, align='center', label='Histogram', color='red')
axarr[1,1].title.set_text("Histograma R")
axarr[1,1].set_xlabel("Intensidade")
axarr[1,1].set_ylabel("Ocorrencias")
axarr[1,2].bar(np.arange(255), img1GHist, width=1, bottom=None, align='center', label='Histogram', color='green')
axarr[1,2].title.set_text("Histograma G")
axarr[1,2].set_xlabel("Intensidade")
axarr[1,2].set_ylabel("Ocorrencias")
axarr[1,3].bar(np.arange(255), img1BHist, width=1, bottom=None, align='center', label='Histogram', color='blue')
axarr[1,3].title.set_text("Histograma B")
axarr[1,3].set_xlabel("Intensidade")
axarr[1,3].set_ylabel("Ocorrencias")
axarr[2,0].imshow(img2RGB)
axarr[2,0].title.set_text("Imagem Original")
axarr[2,1].imshow(img2R)
axarr[2,1].title.set_text("Imagem Canal R")
axarr[2,2].imshow(img2G)
axarr[2,2].title.set_text("Imagem Canal G")
axarr[2,3].imshow(img2B)
axarr[2,3].title.set_text("Imagem Canal B")
axarr[3,0].bar(np.arange(255), img2RHist, width=1, bottom=None, align='center', label='Histogram', color='red', alpha=0.5)
axarr[3,0].bar(np.arange(255), img2GHist, width=1, bottom=None, align='center', label='Histogram', color='green', alpha=0.4)
axarr[3,0].bar(np.arange(255), img2BHist, width=1, bottom=None, align='center', label='Histogram', color='blue', alpha=0.3)
axarr[3,0].title.set_text("Histograma")
axarr[3,0].set_xlabel("Intensidade")
axarr[3,0].set_ylabel("Ocorrencias")
axarr[3,1].bar(np.arange(255), img2RHist, width=1, bottom=None, align='center', label='Histogram', color='red')
axarr[3,1].title.set_text("Histograma R")
axarr[3,1].set_xlabel("Intensidade")
axarr[3,1].set_ylabel("Ocorrencias")
axarr[3,2].bar(np.arange(255), img2GHist, width=1, bottom=None, align='center', label='Histogram', color='green')
axarr[3,2].title.set_text("Histograma G")
axarr[3,2].set_xlabel("Intensidade")
axarr[3,2].set_ylabel("Ocorrencias")
axarr[3,3].bar(np.arange(255), img2BHist, width=1, bottom=None, align='center', label='Histogram', color='blue')
axarr[3,3].title.set_text("Histograma B")
axarr[3,3].set_xlabel("Intensidade")
axarr[3,3].set_ylabel("Ocorrencias")
axarr[4,0].imshow(img3RGB)
axarr[4,0].title.set_text("Imagem Original")
axarr[4,1].imshow(img3R)
axarr[4,1].title.set_text("Imagem Canal R")
axarr[4,2].imshow(img3G)
axarr[4,2].title.set_text("Imagem Canal G")
axarr[4,3].imshow(img3B)
axarr[4,3].title.set_text("Imagem Canal B")
axarr[5,0].bar(np.arange(255), img3RHist, width=1, bottom=None, align='center', label='Histogram', color='red', alpha=0.5)
axarr[5,0].bar(np.arange(255), img3GHist, width=1, bottom=None, align='center', label='Histogram', color='green', alpha=0.4)
axarr[5,0].bar(np.arange(255), img3BHist, width=1, bottom=None, align='center', label='Histogram', color='blue', alpha=0.3)
axarr[5,0].title.set_text("Histograma")
axarr[5,0].set_xlabel("Intensidade")
axarr[5,0].set_ylabel("Ocorrencias")
axarr[5,1].bar(np.arange(255), img3RHist, width=1, bottom=None, align='center', label='Histogram', color='red')
axarr[5,1].title.set_text("Histograma R")
axarr[5,1].set_xlabel("Intensidade")
axarr[5,1].set_ylabel("Ocorrencias")
axarr[5,2].bar(np.arange(255), img3GHist, width=1, bottom=None, align='center', label='Histogram', color='green')
axarr[5,2].title.set_text("Histograma G")
axarr[5,2].set_xlabel("Intensidade")
axarr[5,2].set_ylabel("Ocorrencias")
axarr[5,3].bar(np.arange(255), img3BHist, width=1, bottom=None, align='center', label='Histogram', color='blue')
axarr[5,3].title.set_text("Histograma B")
axarr[5,3].set_xlabel("Intensidade")
axarr[5,3].set_ylabel("Ocorrencias")
```
