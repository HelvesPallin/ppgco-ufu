```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1.0, 5.1, 0.2)
y = np.exp(x)
plt.plot(x, y, 'y*')
plt.title('Meu grfico de exponencial!') 

plt.text( 1, 100, "Texto no ponto 1,100" ) 

plt.show()

x = np.arange(1.0, 5.1, 0.1)
y = 1.2 ** x
x2 = x[:]  
y2 = np.log(x2)
x3 = x
y3 = np.cos(x3)
plt.title('Trs funes')

plt.plot(x, y,'bo', x2, y2, 'y*', x3, y3, 'r--')

plt.legend( ('exponencial', 'logaritmo', 'cosseno'), loc='upper right' )

plt.show()

plt.subplot(1, 3, 1)

pi = np.pi

x1 = np.arange( -4*pi, 4*pi, 0.1)
y1 = np.sin(x1)
plt.title('seno')
plt.plot(x1, y1, 'b--')

plt.subplot(1, 3, 2)
x2 = x1
y2 = np.cos(x2)
plt.title('cosseno')
plt.plot(x2, y2, 'go')

plt.subplot(1,3, 3)
x3 = x1
y3 = np.tan(x3)
plt.plot(x3, y3, 'r')
plt.title('tangente')
plt.show()

fig = plt.figure()

x = np.arange(-10, 10.1, 0.1)
y = x**3 + 4*x + 2

plt.plot( x, y )

plt.axis( [-12, 15, -1000, 1000] )
plt.xlabel( 'meu eixo x' )
plt.ylabel( '\nmeu eixo y' )

fig.savefig( R'.\grafico.png')
plt.show()
```
